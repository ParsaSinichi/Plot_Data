```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
x = np.arange(0,30,0.1)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.ylim(-2,3)
plt.savefig('./plt.jpg')

plt.show()
```

```python
img = cv2.imread('plt.png')
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
cat = cv2.cvtColor(cv2.imread('cat.jpg'),cv2.COLOR_BGR2RGB)
kernel = np.array([
    [2,1,1,2],
    [1,-6,-3,1],
    [1,-3,-6,1],
    [2,1,1,2]
    ])
cat_processed = cv2.filter2D(src=img, ddepth=-1, kernel=kernel)
plt.imshow(cat)
plt.imshow(cat_processed)
```

```python
lb=np.array([0, 0, 0])
ub = np.array([254,100,100])
mask = cv2.inRange(img,lb,ub)
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)
mask = 255*(mask > 230)

plt.imshow(mask)
```

```python
lb=np.array([0, 100, 0])
ub = np.array([254,254,90])
mask = cv2.inRange(img,lb,ub)
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)
mask = 255*(mask > 230)

plt.imshow(mask)
```
